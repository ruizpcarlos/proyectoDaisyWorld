#Depredadores

Agregamos la existencia de animales al mundo. Un hervíboro, conejos, que comen las tres especies de margaritas indistintamente y un carnívoro, zorros, que se alimentan de conejos. 

Modelar la población de animales requiere de varios supuestos; como la proporción de biomasa que acumula cada especie. Esto es, la proporción del alimento que consume un animal que termina como masa del mismo; debido a la energía perdida por respirar. Fijamos también la población mínima de cada animal, notando que no existen restricciones espaciales a su crecimiento como en el caso de las margaritas. Existe también el concepto de disponibilidad de comida, donde se asume que los zorros deben esforzarse más por alimentarse que los conejos. Las plantas son estáticas y por lo tanto fácilmente disponibles como alimento, pero los conejos deben ser atrapados por los zorros para poder ser comidos. Las tasas de muerte y crecimiento también juegan un papel importante, especialmente pensando que los zorros no tienen depredadores que controlen su población; asumimos que los zorros crecen más lentamente que los conejos.

Las condiciones iniciales que dejamos fijas son:

Conejos:

tasa_cre = 2.5            tasa de crecimiento

const_biomasa = 0.3       constante de transformación de alimento en biomasa

poblacion_min = 0.01      población mínima

disponibilidad = 0.8      tasa de disponibilidad de alimento

Zorros:

tasa_cre = .8

const_biomasa = .5

poblacion_min = 0.01

disponibilidad = 0.4

Hacemos interactuar las tasas de muerte y las poblaciones (tasas) iniciales, para conejos, zorros y margaritas (blancas, negras y grises).




```python
%matplotlib inline
%pylab inline

import numpy as np
import matplotlib.pyplot as plt
from IPython.html.widgets import interact, fixed
from IPython.html import widgets
```

```python
class Planeta:
     ##agregamos a los animales como objetos del mundo
    def __init__(self, plantas, animales, albedo_f, luminosidad, prop_fertil, q = 15, boltz = 5.67e-8):
        self.blanca = plantas[0]
        self.negra = plantas[1]
        self.cheat = plantas[2]
        self.conejo = animales[0]
        self.zorro = animales[1]
        self.albedo_f = albedo_f
        self.luminosidad = luminosidad
        self.boltz = boltz
        self.albedo_p = None
        self.temp_g = []
        self.temp_g_vacio = []
        self.q = q
        self.prop_fertil = prop_fertil
        self.area_blanca = []
        self.area_negra = []
        self.area_cheat = []
        self.pob_conejo = []
        self.pob_zorro = []
        self.step = 0
        
    def actualiza_temp_g(self):
        #cálculo de temperatura local, con el albedo planetario calculado anteriormente
        self.temp_g = np.append(self.temp_g,((self.luminosidad[self.step]/(self.boltz))*(1 - self.albedo_p))**0.25 )
        
    def temp_planeta_vacio(self):
        for i in self.luminosidad:
            self.temp_g_vacio = np.append(self.temp_g_vacio,((i/(self.boltz))*(0.5))**0.25 )
    
    def actualiza_albedo_p(self):
        #actualización del albedo planetario, pensando en un mundo cuadrado
        self.prop_fertil = 1 - self.blanca.prop_area - self.negra.prop_area - self.cheat.prop_area
        self.albedo_p = self.blanca.prop_area*self.blanca.albedo + self.negra.prop_area*self.negra.albedo + (self.prop_fertil + self.cheat.prop_area)*self.albedo_f

    def vivir(self):
        for i in self.luminosidad:
            #actualizacion del mundo lineal, cambian las plantas, los conejos comen plantas y los zorros conejos; en ése orden
            self.actualiza_albedo_p()
            self.actualiza_temp_g()
            self.blanca.actualizar(self)
            self.negra.actualizar(self)
            self.cheat.actualizar(self)
            self.conejo.actualiza_conejo(self)
            self.zorro.actualiza_zorro(self)
            #guardamos los resultados para graficar
            self.area_blanca = np.append(self.area_blanca,self.blanca.prop_area)
            self.area_negra = np.append(self.area_negra,self.negra.prop_area)
            self.area_cheat = np.append(self.area_cheat,self.cheat.prop_area)
            self.pob_conejo = np.append(self.pob_conejo, self.conejo.poblacion)
            self.pob_zorro = np.append(self.pob_zorro, self.zorro.poblacion)
            self.step += 1
    
    def graficar_mundo(self):
        self.temp_planeta_vacio()
        grafica_mundo, ax = plt.subplots(3, sharex = True,figsize=(14,20))
        ax[0].plot(self.luminosidad, self.area_blanca, linewidth = .75)
        ax[0].plot(self.luminosidad, self.area_negra, color = "black", linewidth = .75)
        ax[0].plot(self.luminosidad, self.area_cheat, color  = "grey", linewidth = .75)
        ax[0].plot(self.luminosidad, self.area_blanca + self.area_negra + self.area_cheat, color = "red", linewidth = .5)
        ax[0].set_ylabel("Proporcion de area cubierta")
        ax[0].legend(["Area blanca","Area negra","Area gris","Area total"])
        ax[1].plot(self.luminosidad,self.pob_conejo, color = "blue", linewidth = 1.5, label="Conejos")
        ax[1].plot(self.luminosidad,self.pob_zorro, color = "red", linewidth = 1.5, label = "Zorros")
        ax[1].legend(["Conejos","Zorros"])
        ax[1].set_ylabel("Poblacion")
        ax[2].plot(self.luminosidad,self.temp_g, color = "blue", linewidth = 1.5, label = "DaisyWorld")
        ax[2].plot(self.luminosidad, self.temp_g_vacio, color = "red", linewidth = 1.5, label = "Mundo vacio")
        ax[2].legend(["DaisyWorld","Mundo Vacio"])
        ax[2].set_ylabel("Temperatura global (K)")
        ax[2].set_xlabel("Luminosidad solar")
        
```

```python
##sin cambios
class Planta:
    dif_area =0
    temp_l=0
    tasa_cre = 0
    def __init__(self, color, albedo, prop_area, tasa_muerte = 0.1, superior = 313, inferior = 278):
        self.color = color
        self.albedo = albedo
        self.prop_area = prop_area
        self.tasa_muerte = tasa_muerte
        self.superior = superior
        self.inferior = inferior
           
    def actualiza_temp_l(self, mundo):
        self.temp_l = mundo.q * (mundo.albedo_p - self.albedo) + mundo.temp_g[mundo.step]
        
    def actualiza_taza_cre(self):
        if self.temp_l > self.inferior and self.temp_l < self.superior :
            self.tasa_cre = (self.temp_l - self.inferior) * (self.superior - self.temp_l) * 4 / (self.inferior -self.superior)**2
        else:
            self.tasa_cre = 0          
            
    def actualiza_area(self,mundo):
        self.dif_area = self.prop_area*(self.tasa_cre * mundo.prop_fertil - self.tasa_muerte)
        self.prop_area = self.prop_area  + self.dif_area
        if self.prop_area <= 0.01:
            self.prop_area = 0.01
        if self.prop_area >= .99:
            self.prop_area = .99

    def actualizar(self,mundo):
        self.actualiza_temp_l(mundo)
        self.actualiza_taza_cre()
        self.actualiza_area(mundo)
```

```python
##agregamos la clase animal, con métodos distintos de actualización de población, de acuerdo a las ecuaciones planteadas por Mark Wittwer

class animal:
    #incluimos a las plantas como atributos, para la interaccion entre todos los habitantes planetarios
    def __init__(self, plantas, poblacion, poblacion_min, tasa_muerte, disponibilidad, tasa_cre, const_biomasa):
        self.blanca = plantas[0]
        self.negra = plantas[1]
        self.cheat = plantas[2]
        self.poblacion = poblacion #poblacion inicial, interactiva
        self.poblacion_min = poblacion_min
        self.tasa_muerte = tasa_muerte #interactiva
        self.disponibilidad = disponibilidad
        self.tasa_cre = tasa_cre
        self.const_biomasa = const_biomasa
    
        
    def actualiza_conejo(self, mundo): 
        #actualizacion de pobacion
        self.dif = self.tasa_cre *self.poblacion*((self.disponibilidad*(mundo.blanca.prop_area + mundo.negra.prop_area + mundo.cheat.prop_area))**2 - self.tasa_muerte);
        self.poblacion = self.poblacion + self.const_biomasa*self.dif
        #restamos las plantas comidas por los conejos
        if self.dif >0:
            mundo.blanca.prop_area -= mundo.blanca.prop_area * self.dif
            mundo.negra.prop_area -= mundo.negra.prop_area * self.dif
            mundo.cheat.prop_area -= mundo.cheat.prop_area * self.dif
        #aseguramos que el mínimo de población se conserve
        if self.poblacion < self.poblacion_min:
            self.poblacion = self.poblacion_min

                
    def actualiza_zorro(self, mundo): 
        #actualizacion de pobacion
        self.dif = self.tasa_cre*self.poblacion*((self.disponibilidad*mundo.conejo.poblacion)**2 - self.tasa_muerte);
        self.poblacion = self.poblacion + self.const_biomasa*self.dif
        #restamos las plantas comidas por los conejos
        if self.dif >0:
            mundo.conejo.poblacion -= mundo.conejo.poblacion * self.dif
        #aseguramos que el mínimo de población se conserve
        if self.poblacion < self.poblacion_min:
            self.poblacion = self.poblacion_min

        
```

```python
from IPython.html.widgets import interact, fixed
from IPython.html import widgets

def interactivo (area_inicial_blanca,area_inicial_negra,area_inicial_gris, pob_inicial_conejos, pob_inicial_zorros, tasa_muerte_blanca,tasa_muerte_negra,tasa_muerte_gris, tasa_muerte_conejos, tasa_muerte_zorros, ):
    ##nos aseguramos que los valores aportados por el usuario cumplan los supuestos y condiciones del modelo
    assert 0 < area_inicial_blanca + area_inicial_negra + area_inicial_gris < 1 
    assert 0< tasa_muerte_blanca < 1
    assert 0< tasa_muerte_negra < 1
    assert 0< tasa_muerte_gris < 1
    area_f = 1 - (area_inicial_blanca + area_inicial_gris + area_inicial_negra)
    
    print "Total de area cubierta por plantas = " + str(area_inicial_blanca +area_inicial_negra + area_inicial_gris) + " (< 1 para que corra)"

    #creamos los objetos con los valores aportados
    negra = Planta("negra",.25, area_inicial_negra)
    cheat = Planta("cheat", 0.5, area_inicial_gris)
    blanca = Planta("blanca",.75, area_inicial_blanca)

    ##Se hizo una simulación de 1000 pasos, a diferencia de los modelos anteriores, para ver con más precisión la evolucion del sistema
    luminosidad = np.linspace(.6, 1.8, 1000) 
    luminosidad *= 1000
    plantas = [blanca, negra, cheat]

    conejo = animal(plantas, poblacion = pob_inicial_conejos, poblacion_min = .01, tasa_muerte = tasa_muerte_conejos, disponibilidad = .8, tasa_cre =  2.5, const_biomasa = .3)
    zorro = animal(plantas, poblacion = pob_inicial_zorros, poblacion_min = .005, tasa_muerte = tasa_muerte_zorros, disponibilidad = .4, tasa_cre = .8, const_biomasa = .4)

    animales = [conejo, zorro]
    
    #corremos el modelo
    mundo_cruel = Planeta(plantas, animales, 0.5, luminosidad, area_f)
    mundo_cruel.vivir()
    mundo_cruel.graficar_mundo()
```

#Conclusión

Después de interactuar extensivamente con nuestro modelo completo, que incluye depredadores, observamos que el equilibrio biológico del planeta es bastante robusto. El equilibrio térmico se mantiene para un amplio espectro de luminosidades solares, a pesar de la competencia por espacio de las margaritas grises que no contribuyen al albedo planetario y de la depredación que los conejos representan a las plantas. La vida en el planeta sobrevive en los mismos rangos que en el modelo sencillo de margaritas blancas y negras; son estas la base del ecosistema completo, una vez que ni siquiera una proporción de área cercana a uno de flores blancas puede mantener la temperaura en un rango habitable, todo el planeta muere. 

Gracias a la posibiliad de incluir muchas actualizaciones del modelo (mil), vemos oscilaciones en las gráficas que hablan de un equilibrio biológico no antes visto en previas reproducciones de Daisyworld. Tanto plantas como animales oscilan recíprocamente, representando la competencia por sobevivir y el equilibrio que esto genera. La temperatura tambié oscila por lo tanto pero dentro de rangos aceptables para la vida.

Esta oscilación térmica refuerza particularmente la teoría de *Gaia*, pues el planeta también se comporta como un sistema cambiante. Este mundo cambia dependiendo cómo cambien sus habitantes; estos, en su interacción aseguran las condiciones para sus supervivencia conjunta a mediano plazo.
